Sendo assim, pela distribuição dos histogramas, é possível verificar indícios da presença de valores atípicos, como nas variáveis price e minimum_nights.

####  **Q3.2 Histogramas sem *Outliers***

Visto que foram identificados *outliers* nas variáveis `price` e `minimum_nights`, será aplicada a Regra dos 1,5 * IQR para tratar os dados. Essa regra é uma abordagem comumente usada para identificar outliers dentro de um *dataset*.

O cálculo dessa regra é feito da seguinte forma:

 - $ Limite Superior = Q3 + 1.5 * IQR $

 - $ Limite Inferior = Q1 - 1.5 * IQR $
  - em que IQR representa a diferença entre o terceiro quartil (Q3) e o primeiro quartil (Q1) do *dataset*.

No caso, será calculado apenas o limite superior. Posteriormente, será realizado a limpagem do DataFrame e plotado os histogramas das variáveis.

```
# Cálculo IQR para variável 'price'
q1_price = df.price.quantile(.25)
q3_price = df.price.quantile(.75)
IQR_price = q3_price - q1_price
print(f'Valor 1º Quartil: {q1_price:.2f} dólares')
print(f'Valor 3º Quartil: {q3_price:.2f} dólares')
print(f'Valor do IQR: {IQR_price:.2f} dólares')

# Limite para considerar outlier
sup_price = q3_price + (1.5 * IQR_price)
print(f'Valor Limite Superior: {sup_price:.2f} dólares')
```

```
# Cálculo IQR para variável 'minimum_nights'
q1_minimum_nights = df.minimum_nights.quantile(.25)
q3_minimum_nights = df.minimum_nights.quantile(.75)
IQR_minimum_nights = q3_minimum_nights - q1_minimum_nights
print(f'Valor 1º Quartil: {q1_minimum_nights:.0f} noite')
print(f'Valor 3º Quartil: {q3_minimum_nights:.0f} noites')
print(f'Valor do IQR: {IQR_minimum_nights:.0f} noites')

# Limite para considerar outlier
sup_minimum_nights = q3_minimum_nights + (1.5 * IQR_minimum_nights)
print(f'Valor Limite Superior: {sup_minimum_nights:.0f} noites')
```

### **Q4. Qual o tipo de imóvel mais alugado no Airbnb?**

A coluna `room_type` refere-se ao tipo de acomodação disponível para aluguel anunciado no Airbnb. Essa variável indica o estilo ou categoria do espaço que está sendo oferecido, podendo ser:
 - *Entire home/apt* (casa/apartamento inteiro)
 - *Private room* (quarto privativo)
 - *Shared room* (quarto compartilhado)
 - *Hotel room* (quarto de hotel)
 - Ou outra categoria definida pelo Airbnb

```
#@title
# Definir as cores para o gráfico
azul1, azul2, azul3 = "#253760", "#6497b1", "#b3cde0"

# Porcentagem de cada tipo de imóvel disponível
porcentagem = df_clean.room_type.value_counts() / df_clean.shape[0]

# Agrupar as categorias "Hotel room" e "Shared room"
porcentagem['Outros'] = porcentagem['Hotel room'] + porcentagem['Shared room']
del porcentagem['Hotel room']
del porcentagem['Shared room']

# Definir as legendas das categorias
imoveis = porcentagem.index.tolist()
imoveis[-1] = 'Hotel room + \nShared room'

# Plotar o gráfico de pizza com as cores definidas
fig, ax = plt.subplots(figsize=(6, 4))
fatias, legendas, rotulos = ax.pie(porcentagem, labels=imoveis, autopct='%1.1f%%')

# Definir cores das categorias
fatias[0].set_facecolor(azul1)
fatias[1].set_facecolor(azul2)
fatias[2].set_facecolor(azul3)

# Definir as cores e estilos das legendas
legendas[0].set_color(azul1)
legendas[0].set_fontweight('bold')
legendas[1].set_color(azul2)
legendas[1].set_fontweight('bold')
legendas[2].set_color(azul3)
legendas[2].set_fontweight('bold')

# Definir as cores dos rótulos de dados
for rotulo in rotulos:
    rotulo.set_color('white')

ax.set_title('Porcentagem dos Tipos de Imóveis', color=azul1, fontsize=14, fontweight='bold')

fig.tight_layout()
plt.show()
```

##### ***NOTA***
Apesar de gráficos de pizza não serem recomendados para visualização de dados, optei por utilizá-lo devido às quantidades de imóveis denominadas 'Hotel Room' e 'Shared Room'. Essas categorias representam um valor muito baixo na porcentagem total dos imóveis do Airbnb e, portanto, não eram visíveis no gráfico de barras.
